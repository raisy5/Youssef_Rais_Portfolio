###**Vecteur**
-------
Dans cette section, nous allons passer en revue les méthodes les plus importantes qui nous permettront d'accélérer notre progression et de simplifier nos **calculs algébriques**. Chaque méthode sera accompagnée d'un commentaire expliquant son rôle et son utilité.

Ces outils nous aideront à optimiser notre travail, à améliorer notre compréhension des concepts mathématiques et à mieux **manipuler les structures de données**.

```python
vecteur=np.array([1,2,3])
vecteur #construction de vecteur


print(vecteur.shape) #dimension du vecteur(retourne un tuple )
print(vecteur.size) #nombre d'éléments du vecteur
print(vecteur.dtype) #type des éléments du vecteur
print(vecteur.astype(float) )#conversion du type des éléments du vecteur
print(vecteur.tolist()) #conversion du vecteur en liste
print(vecteur.reshape(3,1)) #redimensionnement du vecteur
print(vecteur.transpose()) #transposition du vecteur
print(vecteur.sum()) #somme des éléments du vecteur
print(vecteur.mean()) #moyenne des éléments du vecteur
print(vecteur.std()) #écart-type des éléments du vecteur
print(vecteur.var()) #variance des éléments du vecteur
print(vecteur.min())#l'élément le plus petit du vecteur
print(vecteur.max())#l'élément le plus grand du vecteur
print(vecteur.argmin())#l'indice de l'élément le plus petit du vecteur
print(vecteur.argmax())#l'indice de l'élément le plus grand du vecteur
print(np.zeros(4))#vecteur de zero
print(np.linspace(0,10,5))#vecteur de 5 éléments entre 0 et 10
print(np.arange(0,10,2))#vecteur de 0 à 10 avec un pas de 2
print(np.dot(vecteur, vecteur))#produit scalaire entre deux vecteurs

#La selection d'un élément du vecteur se fait de la meme manière que pour les listes
print("Selection d'element " )
print(vecteur[0])
print(vecteur[-1])
print(vecteur[0:2])
print(vecteur[::2])
print(vecteur[::-1])
```

###Les matrices avec numpy
-----

```python
#Les méthodes des vecteurs sont valables pour les matrices

#Création de Matrice
matrice=np.array([[1,2,1],[4,5,6],[7,8,9]])
print(matrice)

print(matrice.shape) #dimension de la matrice(retourne un tuple )
print(matrice.ndim) #nombre de dimension de la matrice(rang de la matrice)

#Opérations sur les matrices

print(matrice.transpose()) #transposition de la matrice
print(matrice.sum()) #somme des éléments de la matrice
print(matrice.sum(axis=0)) #somme des éléments de chaque colonne
print(matrice.sum(axis=1)) #somme des élément de chaque ligne (axis(1 ou 0) applicable pour les moyenne pour les ecart type aussi )
print(np.dot(matrice, matrice)) #produit matriciel entre deux matrices
print(np.linalg.det(matrice)) #déterminant de la matrice

# Calcul de l'inverse
print(np.linalg.inv(matrice)) #inverse de la matrice,attention au matrice singulière
print("Valeurs propres ",np.linalg.eig(matrice).eigenvalues) #valeurs propre de la matrice
print("Vecteurs propres ",np.linalg.eig(matrice).eigenvectors) #vecteurs propre de la matrice
```

#Modèle Linéaire simple
-----
**Méthode** :Nous utiliserons [ces données](https://www.kaggle.com/code/shubham47/linear-regression-salary-dataset/input) pour implémenter un modèle de régression linéaire simple. Ensuite, nous afficherons la droite de prédiction superposée au nuage de points afin de visualiser la relation entre les variables.

```python
data=pd.DataFrame(pd.read_csv('Salary_Data.csv'))
```

##Première étape :
+ Utilisation du fichier CSV Salary_Data stocké dans le répertoire data.
+ Application des méthodes de statistique descriptive pour estimer l'équation de la droite de régression et ainsi modéliser la relation entre les variables

+ Rappelons que : y =b0+b1*x (forme linéaire simple)

+ On va expliquer le salaire par le nombre d'années d'experience
++ **variable explicative :x=Année d'expérience**
++*texte en italique* **variable à expliquer :y=Salaire**

+ Donc comment calculer b1 et b0
++ b1=cov(x,y)/var(x)
++ b0=Moyenne(y)-b1*moyenne(x)

<font color='red'>
Dans le bloc précécent, corriger

- "le nombre d'annéeS d'expérience"
- "Donc comment calculé"... en "Comment calculer"
</font>

<font color='green'>Fait</font>

```python
import matplotlib.pyplot as plt
#on va utilisé mathplotlib pour afficher le nuage des points

x=data.YearsExperience
y=data.Salary

b1=np.cov(x,y)[0,1]/np.var(x)#comme cov(x,y)est la matrice de covariance on prend l'élément [0,1]
print('b1 : ',b1)
b0=np.mean(y)-b1*np.mean(x)
print('b0 : ',b0)

#Nous avons calculé les deux coefficients b1 et b2
#nous allons utilisé ces coefficient pour calculé l'estimations de chaque valeur y
estim=b0+b1*x
print('estimations : ',estim)

#Affichage des données avec l'estimation(la droite de regression linéaire)
data.plot(kind='scatter', x='YearsExperience', y='Salary', alpha=0.5)
plt.plot(x, estim, color='red')
plt.xlabel('Année d\'expérience')
plt.ylabel('Salaire')
plt.title('Modèle linéaire simple')
plt.show()
```

#Modèle Linéaire multiples
----
Nous allons utiliser un échantillon de [données](https://www.kaggle.com/code/racksonsit/multiple-linear-regression-price-dataset/input?select=Date-Wise-Prices-all-Commodity.xlsx) provenant de Kaggle pour illustrer un modèle linéaire multiple en classe. Cette approche nous permettra de mieux comprendre son fonctionnement et d'analyser ses performances à partir de données réelles.


```python

#importation du fichier word
data=pd.read_excel('Date-Wise-Prices-all-Commodity.xlsx')
print(len(data))
#y.plot.box()#on remarque que la majorité des prix sont en dessous de 10000
#donc nous allons diminuer notre dataset en prenons que les éléments inférieur à 10000
data=data[data.Min_x0020_Price<=10000]
print(len(data))

#####Nouvelle fonctionnalité sklearn
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
# Transformation des données catégoriques
#en données numériques exploitables pour le modèle
data["Market"]=lbl.fit_transform(data["Market"])
print(data["Market"])#on remarque qu'on vient de transformer nos données en des données numérique utilisable pour le modèle
data["Commodity"]=lbl.fit_transform(data["Commodity"])
data["Grade"]=lbl.fit_transform(data["Grade"])
data["Variety"]=lbl.fit_transform(data["Variety"])

print('---------------------------------------')
print(data)
print('---------------------------------------')
print('    ')
```

## Exemple suivant la méthode vue en cours
-----


```python
#J'ai choisi de garder que deux variables explicatives afin de proposer une illustration 3 dimension ensuite
data1=data[data.columns.difference(['diffgr:id', 'msdata:rowOrder', 'State', 'District',
        'Variety', 'Grade', 'Arrival_Date',
       'Max_x0020_Price', 'Modal_x0020_Price'])]
print(data1)

#Nous allons essayé d'expliquer le prix minimum
#par deux variables explicatives : le market et commodity

#Notre variable à expliquer
Y=data1.Min_x0020_Price
Y=Y[Y<=10000]

print(Y)
#On va devoir supprimer celle ci de notre jeux de données
#pour garder que les variables explicatives
data1=data1.drop(columns=['Min_x0020_Price'])

print(ma_serie.values)
#Vecteur de la variable à expliquer
Y=Y.values.reshape(-1,1)

print(data1)
```

```python
print("Y: ", Y)
data1.plot(kind='scatter',x='Market',y='Commodity')
plt.xlabel('Market')
plt.ylabel('Commodity')

#Matrice des variables explicatives
X=data1.values
#print('X:',X)

#ajouter une colonne de 1 à X
X=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
print('X :',X)

Beta=np.linalg.inv(X.T@X)@X.T@Y #@ pour la multiplication matricielle
#nous avons obtenu notre vecteur de Beta ,donc estimation
print('vecteur de Beta : ', Beta)
estim=X@Beta
print('estimations : ',estim)
#R2
from sklearn.metrics import r2_score
r2 = r2_score(Y, estim)
print("R2 :",r2)#on peut voir que le modèle n'est pas performant
#reste à afficher les données avec leur estimation en  3 dimension



```

## Remarque
----
Notre modèle n'est pas bien adapté aux données, car il n'existe pas de relation de linéarité entre les variables.

Ainsi, nous conserverons cet exemple pour illustrer ce constat. En effet, l'absence de linéarité est visible à travers le graphique et confirmée par un R² très faible (0.04), indiquant que la régression linéaire ne capture pas correctement la structure des données.

Dans ce cas, un modèle de classification, comme un arbre de décision, serait bien mieux adapté, permettant une approche plus efficace pour segmenter et prédire les valeurs en fonction des caractéristiques disponibles

        En régression, la valeur prédite est la moyenne des valeurs des voisins.
